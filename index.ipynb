```python
# STEP 1. Importinh libs
import pandas as pd
import pymongo
import requests
import json
import matplotlib.pyplot as plt
import sqlite3
import numpy as np
from IPython.display import HTML
with open('../key.txt','r') as f:
    sk = f.readline()
```

```python
# STEP2. Class for DarkSky API
class WeatherGetter(object): 
    
    def __init__(self):
        self.secret_key = sk
        self.berlin_lat = "52.5200"
        self.berlin_long = "13.4050"
        self.url_base = "https://api.darksky.net/forecast"
        self.exclude = 'currently,flags,minutely,hourly,alerts'
        
    def get_weather_data_for_date(self, datetime_string, verbose=True):
        
        year, month, day = self.format_datetime(datetime_string)
        
        datetime = "{}-{}-{}T12:00:00".format(year, month, day)
        full_url = "{}/{}/{},{},{}?exclude={}".format(self.url_base, self.secret_key, 
                                                     self.berlin_lat, self.berlin_long, 
                                                     datetime, self.exclude)
        response = requests.get(full_url)
        if response.status_code == 200:
            if verbose:
                print(response.status_code)
            return response
        else: 
            raise ValueError("Error getting data from DarkSky API: Response Code {}".format(response.status_code))
            
    def was_raining(self, response, verbose=True):
        data = json.loads(response.text)
        daily = data['daily']
        data =  daily['data']
        data = data[0]
        if data['icon'] == 'rain':
            if verbose:
                print(data['icon'])
            return True
        else:
            if verbose:
                print(data['icon'])
            return False
    
    def format_datetime(self, datetime_string):
        year = datetime_string[:4]
        month = datetime_string[5:7]
        day = datetime_string[8:]
        
        return year, month, day
    
    def did_rain_on_date(self, datetime_string):
        
        response = self.get_weather_data_for_date(datetime_string, verbose=False)
        did_rain = self.was_raining(response, verbose=False)
        
        return did_rain
    
    def get_weather_for_all_dates(self, dates_list):
        """Expects input of dates in yyyy-mm-dd format
        
        Returns a dictionary where each date is the key. Rain days have a value of True, all others are False"""
        
        weather_dict = {}
        
        for date in dates_list:
            weather_dict[date] = self.did_rain_on_date(date)
        
        return weather_dict
```

```python
# STEP 3. Connection to DB and collecting rain-data (one-time-request)
conn = sqlite3.connect('database.sqlite')
cursor = conn.cursor()

def sql_to_df(SQL_CMD, cursor=cursor):
    results = cursor.execute(SQL_CMD).fetchall()
    df = pd.DataFrame(results)
    df.columns = [i[0] for i in cursor.description]
    return df
req_dates_2011 = '''SELECT Date FROM Matches WHERE Season=2011'''
match_dates = list(set(list(sql_to_df(req_dates_2011)['Date'])))

rainy_days = dict()
for i in match_dates:
    wg = WeatherGetter()
    rainy_days[i] = wg.did_rain_on_date(i)
```

```python
# STEP 4. Creating DATAFRAME
req_all_2011 = '''SELECT * FROM Matches WHERE Season=2011'''
initial_DF = sql_to_df(req_all_2011)
rainy_matches = []

for i in initial_DF['Date']:
    if i in rainy_days.keys():
        rainy_matches.append(rainy_days[i])

rainy_matches = pd.Series(rainy_matches)
df_len = len(rainy_matches)
initial_DF['Rain'] = rainy_matches

initial_DF = initial_DF.drop(columns=['Match_ID', 'Div', 'Season','Date'], axis=1)
```

```python
# STEP 5. Class for TEAM
class TEAM:
    
    def __init__(self, team_name, df):
        self.team_name = team_name
        self.df = df
        self.df_home = self._calc_home_games()
        self.df_away = self._calc_away_games()
        
        self.team_row = {'team': self.team_name, 
                    'goals': int(self._calc_total_goals()),
                    'wins': int(self._calc_total_wins()), 
                    'loose': int(self._calc_total_loose()),
                    'draw':int(self._calc_total_draw()),
                    'win_percent_on_rain':float(self._calc_win_percent_on_rain()), 
                    'hist':self._save_chart()}
        self.data = dict()

    def _calc_home_games(self):
        self.df_home = self.df.loc[self.df['HomeTeam']==self.team_name]
        return self.df_home
    
    def _calc_away_games(self):
        self.df_away = self.df.loc[self.df['AwayTeam']==self.team_name]
        return self.df_away
    
    def _calc_total_goals(self):
        self.total_goals = self.df_home['FTHG'].sum() + self.df_away['FTAG'].sum()
        return self.total_goals
    
    def _calc_total_wins(self):
        self.total_wins = self.df_home['FTR'].loc[self.df_home['FTR'] == "H"].count()+\
                          self.df_away['FTR'].loc[self.df_away['FTR'] == "A"].count() 
        return self.total_wins
   
    def _calc_total_loose(self):
        self.total_wins = self.df_home['FTR'].loc[self.df_home['FTR'] == "A"].count()+\
                          self.df_away['FTR'].loc[self.df_away['FTR'] == "H"].count() 
        return self.total_wins
    
    def _calc_total_draw(self):
        self.total_wins = self.df_home['FTR'].loc[self.df_home['FTR'] == "D"].count()+\
                          self.df_away['FTR'].loc[self.df_away['FTR'] == "D"].count() 
        return self.total_wins
    
    
    def _calc_win_percent_on_rain(self):
        self.rainy_df_home = self.df_home.loc[self.df_home['Rain'] == True]
        self.rainy_df_away = self.df_away.loc[self.df_away['Rain'] == True]
        self.rainy_wins = self.rainy_df_home['FTR'].loc[self.rainy_df_home['FTR'] == "H"].count()+\
                          self.rainy_df_away['FTR'].loc[self.rainy_df_away['FTR'] == "A"].count() 
        
        try:
            self.total_rains = len(self.rainy_df_home)+len(self.rainy_df_away)
            if self.total_rains != 0:
                return self.rainy_wins / self.total_rains
            else:
                return 0
        except Exception:
            return 0
            
    def _save_chart(self):
        self.path = f'img/{self.team_name}.png'
        
        
        with open(self.path, 'wb') as f:
            self.data = {'wins':self._calc_total_wins(),
                        'loose':self._calc_total_loose(),
                        'draw':self._calc_total_draw()}
            self.chart = plt.pie(self.data.values(), labels = self.data.keys())
            plt.savefig(self.path, transparent=True)
            plt.show()
            self.title = f'{self.team_name} chart'
            self.img_path = [self.title, '<img src="{}"/>'.format(self.path)]
            return self.img_path
```

```python
# STEP 6. CREATING list with TEAM instances (all charts will be rendered as output)
team_list = list(set(initial_DF['HomeTeam']))
teams = []
for i in team_list:
    team = TEAM(i, initial_DF)
    teams.append(team)
```

```python
# STEP 7. Creating Mongo connection and collections
myclient = pymongo.MongoClient("mongodb://127.0.0.1:27017/")
mydb_1 = myclient["db_soccer_1"]
mycol = mydb_1['M2_FINAL_LAB']

for team in teams:
    added_team = mycol.insert_one(team.team_row)
print(myclient.list_database_names())
```
